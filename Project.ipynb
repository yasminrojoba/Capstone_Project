```python
# assign relevant part of JSON to venues
venues = location_results['response']['venues']

# tranform venues into a dataframe
dataframe = pd.json_normalize(venues)
dataframe.head()
```

### Use KMeans Clustering to cluster the dataset and find a suitable location

```python
from sklearn.cluster import KMeans
```

```python
Memphis_Indian_Restaurent = folium.Map(location=[latitude, longitude], zoom_start=12)

X = dataframe['location.lat']
Y = dataframe['location.lng']
Z = np.stack((X, Y), axis=1)

kmeans = KMeans(n_clusters=5, random_state=0).fit(Z)

clusters = kmeans.labels_
colors = ['blue', 'green', 'red', 'yellow','pink']
dataframe ['Cluster'] = clusters

for latitude, longitude, cluster in zip(dataframe['location.lat'], dataframe['location.lng'], dataframe['Cluster']):
    label = folium.Popup(cluster, parse_html=True)
    folium.CircleMarker(
        [latitude, longitude],
        radius=5,
        popup=label,
        color='black',
        fill=True,
        fill_color=colors[cluster],
        fill_opacity=0.7).add_to(Memphis_Indian_Restaurent)  

Memphis_Indian_Restaurent
```

```python
Memphis_Indian_Restaurent.save('map.html')
```
